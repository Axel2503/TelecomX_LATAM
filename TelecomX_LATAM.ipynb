#📌 Extracción

```python
import pandas as pd
url_api="https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/main/TelecomX_Data.json"
df_telecom=pd.read_json(url_api)
print("¡Datos cargados con éxito!")
print(df_telecom.head())
```

#🔧 Transformación

```python
df_telecom.info()

#Conocer el conjunto de datos
import matplotlib.pyplot as plt
conteo_churn=df_telecom['Churn'].value_counts()
print("Conteo exacto de clientes")
print(Conteo_churn)
print("-"*30)
conteo_churn.plot(kind='bar', color=['skyblue', 'salmon'], edgecolor='black')
plt.title('Proporción de Clientes que dejaron la empresa (Churn)')
plt.xlabel('¿Dejó la empresa?')
plt.ylabel('Cantidad de Clientes')
plt.xticks(rotation=0)
plt.show()
```

```python
#Busqueda de valores ausente y duplicados
print("Valores ausentes por columna:")
print(df_telecom.isnull().sum())
print("-" * 30)
print("Cantidad de filas duplicadas:")
print(df_telecom.duplicated().sum())
print("-" * 30)
```

```python
#Errores de formato
df_telecom['Charges.Total'] = pd.to_numeric(df_telecom['Charges.Total'], errors='coerce')
print("Valores nulos en Charges.Total después de corregir el formato:")
print(df_telecom['Charges.Total'].isnull().sum())
```

```python
#Inconsistencias en las categorías
print("Categorías en la columna de Género:")
print(df_telecom['gender'].unique())
print("-" * 30)
print("Categorías en la columna de Contrato:")
print(df_telecom['Contract'].unique())
```

```python
#Manejo de incosistencias
df_telecom = df_telecom.drop_duplicates()
df_telecom['Charges.Total'] = df_telecom['Charges.Total'].fillna(0)
df_telecom = df_telecom.drop(columns=['customerID'])
columnas_servicios = ['OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']

for columna in columnas_servicios:
    df_telecom[columna] = df_telecom[columna].replace('No internet service', 'No')

df_telecom['MultipleLines'] = df_telecom['MultipleLines'].replace('No phone service', 'No')

print("¡Limpieza terminada!")
print("Valores nulos restantes en todo el dataset:", df_telecom.isnull().sum().max())
```

```python
#Columna de cuentas diarias
df_telecom['Cuentas_Diarias'] = df_telecom['Charges.Monthly'] / 30

df_telecom['Cuentas_Diarias'] = df_telecom['Cuentas_Diarias'].round(2)

print("Verificación de la nueva columna 'Cuentas_Diarias':")
print(df_telecom[['Charges.Monthly', 'Cuentas_Diarias']].head())
```

```python
#Estandarización y transformación de datos
traduccion_columnas = {
    'Churn': 'Evasion',
    'gender': 'Genero',
    'SeniorCitizen': 'Adulto_Mayor',
    'Partner': 'Pareja',
    'Dependents': 'Dependientes',
    'tenure': 'Meses_Contrato',
    'PhoneService': 'Servicio_Telefonico',
    'MultipleLines': 'Multiples_Lineas',
    'InternetService': 'Servicio_Internet',
    'OnlineSecurity': 'Seguridad_En_Linea',
    'OnlineBackup': 'Respaldo_En_Linea',
    'DeviceProtection': 'Proteccion_Dispositivo',
    'TechSupport': 'Soporte_Tecnico',
    'StreamingTV': 'Streaming_TV',
    'StreamingMovies': 'Streaming_Peliculas',
    'Contract': 'Contrato',
    'PaperlessBilling': 'Facturacion_Electronica',
    'PaymentMethod': 'Metodo_Pago',
    'Charges.Monthly': 'Cargo_Mensual',
    'Charges.Total': 'Cargo_Total'
}

df_telecom = df_telecom.rename(columns=traduccion_columnas)

df_telecom = df_telecom.replace({'Yes': 1, 'No': 0})

df_telecom['Genero'] = df_telecom['Genero'].replace({'Female': 'Femenino', 'Male': 'Masculino'})

print("¡Datos estandarizados, traducidos y listos para analizar!")
print(df_telecom.head())
```

#📊 Carga y análisis

```python
# Análisis descriptivo
resumen_estadistico = df_telecom.describe()
print("Análisis Descriptivo de los Clientes:")
print(resumen_estadistico)
```

```python
#Distribución de evasión
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

sns.countplot(data=df_telecom, x='Evasion', palette=['skyblue', 'salmon'], ax=axes[0])
axes[0].set_title('Conteo de Clientes (0 = Se queda, 1 = Se va)', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Estado de Evasión')
axes[0].set_ylabel('Cantidad de Clientes')

for p in axes[0].patches:
    axes[0].annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                     ha='center', va='bottom', fontsize=12, color='black', xytext=(0, 5),
                     textcoords='offset points')

conteo = df_telecom['Evasion'].value_counts()

axes[1].pie(conteo, labels=['Se queda (0)', 'Se va (1)'], autopct='%1.1f%%',
            colors=['skyblue', 'salmon'], startangle=90, explode=(0, 0.1), shadow=True)
axes[1].set_title('Proporción de Evasión (%)', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
#Recuento de evasión por variables categóricas
import matplotlib.pyplot as plt
import seaborn as sns

variables_categoricas = ['Genero', 'Contrato', 'Metodo_Pago']

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

for i, variable in enumerate(variables_categoricas):

    sns.countplot(data=df_telecom, x=variable, hue='Evasion', palette=['skyblue', 'salmon'], ax=axes[i])

    axes[i].set_title(f'Evasión por {variable}', fontsize=12, fontweight='bold')
    axes[i].set_xlabel('')
    axes[i].set_ylabel('Cantidad de Clientes')

    axes[i].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
#Conteo de evasión por variables numéricas
import matplotlib.pyplot as plt
import seaborn as sns
variables_numericas = ['Meses_Contrato', 'Cargo_Mensual', 'Cargo_Total']
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

for i, variable in enumerate(variables_numericas):
    sns.boxplot(data=df_telecom, x='Evasion', y=variable, palette=['skyblue', 'salmon'], ax=axes[i])
    axes[i].set_title(f'Distribución de {variable}', fontsize=12, fontweight='bold')
    axes[i].set_xlabel('Estado de Evasión (0 = Se queda, 1 = Se va)')
    axes[i].set_ylabel(variable)
plt.tight_layout()
plt.show()
```

#Introducción
El presente proyecto tiene como objetivo analizar los datos de los clientes de la empresa de telecomunicaciones "Telecom X" para comprender el fenómeno de la evasión o Churn (tasa de cancelación). El Churn es un problema crítico, ya que retener a un cliente siempre es más rentable que adquirir uno nuevo. A través de este análisis, buscamos identificar qué características y comportamientos tienen los clientes que deciden abandonar el servicio, para así poder tomar decisiones informadas que ayuden a la empresa a retenerlos.

#🔹 Limpieza y Tratamiento de Datos
Para asegurar la calidad del análisis, los datos originales pasaron por un riguroso proceso de preparación utilizando Python y Pandas:

Importación: Se extrajo la base de datos en formato JSON directamente desde la API proporcionada.

Manejo de valores nulos y limpieza: Se identificaron valores faltantes en los cargos totales (generalmente de clientes nuevos) y se rellenaron con ceros matemáticos. Además, se eliminó la columna de identificación (customerID) por no aportar valor predictivo.

Ingeniería de características: Se creó una nueva métrica llamada Cuentas_Diarias dividiendo el cargo mensual entre 30, permitiendo una visión del ingreso diario por cliente.

Estandarización: Se tradujeron los nombres de las columnas al español para mayor claridad y se transformaron las respuestas de texto ("Yes"/"No") a un formato binario numérico (1 y 0) para facilitar su procesamiento.

#🔹 Análisis Exploratorio de Datos (EDA)
Durante la exploración visual utilizando matplotlib y seaborn, descubrimos cómo se distribuye la base de clientes:

Distribución del Churn: Se generaron gráficos de barras y de pastel que revelaron la proporción exacta de clientes retenidos frente a los que cancelaron el servicio, mostrando el tamaño real del problema de evasión.

Variables Categóricas: Al cruzar el Churn con variables como el tipo de contrato y el método de pago, visualizamos claramente qué segmentos de clientes son más propensos a irse (por ejemplo, evaluando la diferencia de lealtad entre contratos mensuales vs. anuales).

Variables Numéricas: Mediante diagramas de caja (boxplots), analizamos la relación entre el tiempo de permanencia (Meses_Contrato) y el dinero gastado (Cargo_Mensual) con la decisión de cancelar, observando en qué meses críticos se pierde a la mayoría de los usuarios.

#🔹 Conclusiones e Insights
A partir del análisis de los datos, los hallazgos más destacados son:

El tipo de contrato es clave: Los clientes con contratos de mes a mes presentan una tendencia de evasión significativamente mayor que aquellos con compromisos a largo plazo (1 o 2 años).

Riesgo en los primeros meses: Los diagramas de tiempo de contrato demostraron que la probabilidad de abandono es más alta al inicio de la relación con la empresa. Si el cliente supera esta barrera inicial, tiende a quedarse.

Sensibilidad al precio: Los clientes que cancelan el servicio tienden a tener cargos mensuales más altos en comparación con los que se quedan, lo que indica que el costo es un factor de peso en su decisión.

#🔹 Recomendaciones
Basado en los patrones descubiertos, se sugieren las siguientes estrategias para reducir la evasión:

Incentivar contratos a largo plazo: Ofrecer descuentos especiales, beneficios extra o mejoras de velocidad a los clientes que decidan cambiar su contrato de "mes a mes" a uno anual.

Programa de fidelización temprana: Crear un proceso de acompañamiento o beneficios exclusivos durante los primeros meses de servicio, que es cuando los usuarios corren mayor riesgo de cancelar.

Revisión de tarifas: Analizar los planes con cargos mensuales más altos y ofrecer reestructuraciones o paquetes personalizados antes de que el cliente perciba que el servicio es demasiado costoso y decida irse.
